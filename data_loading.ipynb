```python
# Specify the calendar year to prepare generation, emissions and pricing data
# from. The data is then is saved in parquet files for rapid loading in the
# interface tab. Once you've chosen the year run this cell to load the data,
# this may take up to 10 min.
year = 2021

# The code below automates the data preparation process and should not need to
# be edited by the user.
end_year = year + 1

from pathlib import Path
from datetime import datetime

from getting_data import import_gen_data, import_load_data, \
import_pricing_data, import_emissions_data
import advanced_settings

# Provide artificial start and end dates so we can loop over load later:
start_date = datetime.strptime(f'{year}-01-01 00:00:00', '%Y-%m-%d %H:%M:%S')
end_date = datetime.strptime(f'{end_year}-01-01 00:00:00', '%Y-%m-%d %H:%M:%S')

# ----------------------------- Get Generation Data ----------------------------
gen_data = import_gen_data.get_generation_data(
    advanced_settings.RAW_DATA_CACHE, 
    advanced_settings.GEN_TECH_TYPE_S, 
    start_date=start_date,
    end_date=end_date
)
gen_data.to_parquet(
    advanced_settings.YEARLY_DATA_CACHE / f'gen_data_{year}.parquet'
)

# ------------------------ Get Pricing & Emissions Data ------------------------
emissions_intensity = import_emissions_data.get_avg_emissions_intensity(
    start_date, end_date, advanced_settings.EMISSIONS_CACHE
)
emissions_intensity.to_parquet(
    advanced_settings.YEARLY_DATA_CACHE / f'emissions_data_{year}.parquet'
)

price_data = import_pricing_data.get_wholesale_price_data(
    start_date, end_date,advanced_settings. PRICING_CACHE,
)
price_data.to_parquet(
    advanced_settings.YEARLY_DATA_CACHE /f'price_data_{year}.parquet'
)
```
